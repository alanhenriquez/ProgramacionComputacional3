```python
print("Simulacion de cajero automático")

def TipoDato(number):
  num = None
  for conv in (int, float, complex, str, bool):
    try:
      num = conv(number)
      break
    except ValueError:
      pass

  if num is None:
      return ("Error de entrada")
  else:
      typos = (type(num).__name__);
      return (typos);    


valor = input("Ingrese la cantidad a retirar: ");


def CajeroAutomatico(number):
  
  seguir = 'si';
  while seguir == "si":
    datoPrincipal = number;
    intentos = 0;
    nuevoProceso = 0;
    string1 = " Billetes de "

    while str(TipoDato(datoPrincipal)) != "int":
      intentos += 1
      datoPrincipal = input("Ingrese nuevamente la cantidad a retirar: ");
      if intentos == 2 and str(TipoDato(datoPrincipal)) != "int":
        print("\n" + "Solo se permiten numeros enteros")
      elif intentos > 2 and str(TipoDato(datoPrincipal)) != "int":
        print("\n" + "Recuerde que solo se permiten NUMEROS ENTEROS: 1 y parecidos en adelante... ")

      
    while nuevoProceso == 0:
      print("\n" + "Monto seleccionado: " + str(datoPrincipal) + "$")
      print("Opciones: si o no")
      respuesta = input("Proceder con transaccion? ")
      if respuesta == "si":
        num = int(datoPrincipal)
        dolar20 = 0;
        dolar10 = 0;
        dolar5 = 0;
        dolar1 = 0;

        while num >= 20:
          dolar20 += 1;
          num = num - 20

        while num >= 10:
          dolar10 += 1;
          num = num - 10

        while num >= 5:
          dolar5 += 1;
          num = num - 5

        while num >= 1:
          dolar1 += 1;
          num = num - 1

        while num >= 0.25:
          num = num - 0.25

        while num >= 0.10:
          num = num - 0.10

        while num >= 0.05:
          num = num - 0.05

        while num >= 0.01:
          num = num - 0.01



        print("\n" + "Entregando: ")
        if dolar20 > 0:
          print(str(dolar20) + string1 + "20$");
        if dolar10 > 0:
          print(str(dolar10) + string1 + "10$");
        if dolar5 > 0:
          print(str(dolar5) + string1 + "5$");
        if dolar1 > 0:
          print(str(dolar1) + string1 + "1$");
        print("Total: " + str(datoPrincipal) + "$")
        print("Monto restante: " + str(num) + "$");
        nuevoProceso = 1;


            
      elif respuesta == "no":
        seguir = "no"
      else:
        print("Favor seleccionar entre (si) o (no)")



    while nuevoProceso == 1:
      n = input("¿Quiere realizar un nuevo proceso?")
      nuevaRespuesta = n;

      if nuevaRespuesta == "si":
        datoPrincipal = input("Ingrese nuevo monto a retirar: ");
        print("\n" + "Nuevo monto seleccionado: " + str(datoPrincipal) + "$")
        print("Opciones: si o no")
        respuesta = input("Proceder con transaccion? ")
        if respuesta == "si":
          num = int(datoPrincipal)
          dolar20 = 0;
          dolar10 = 0;
          dolar5 = 0;
          dolar1 = 0;

          while num >= 20:
            dolar20 += 1;
            num = num - 20

          while num >= 10:
            dolar10 += 1;
            num = num - 10

          while num >= 5:
            dolar5 += 1;
            num = num - 5

          while num >= 1:
            dolar1 += 1;
            num = num - 1

          while num >= 0.25:
            num = num - 0.25

          while num >= 0.10:
            num = num - 0.10

          while num >= 0.05:
            num = num - 0.05

          while num >= 0.01:
            num = num - 0.01



          print("\n" + "Entregando: ")
          if dolar20 > 0:
            print(str(dolar20) + string1 + "20$");
          if dolar10 > 0:
            print(str(dolar10) + string1 + "10$");
          if dolar5 > 0:
            print(str(dolar5) + string1 + "5$");
          if dolar1 > 0:
            print(str(dolar1) + string1 + "1$");
          print("Total: " + str(datoPrincipal) + "$")
          print("Monto restante: " + str(num) + "$");
          nuevoProceso = 1;


              
      elif nuevaRespuesta == "no":
        nuevoProceso = 2;
        seguir = "no"
      else:
        print("Favor seleccionar entre (si) o (no)")

  return ("Transaccion terminada")



  

print(CajeroAutomatico(valor));  

```
